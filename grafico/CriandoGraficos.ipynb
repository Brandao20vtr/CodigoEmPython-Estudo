# Criando Gráficos em Python
O conjunto de dados Iris foi usado em R.A. O artigo clássico de Fisher de 1936, The Use of Multiple Measurements in Taxonomic Problems, e também pode ser encontrado no Repositório de Aprendizado de Máquina da UCI.

Inclui três espécies de íris com 50 amostras cada, bem como algumas propriedades sobre cada flor. Uma espécie de flor é linearmente separável das outras duas, mas as outras duas não são linearmente separáveis ​​uma da outra.

**Dataset:** https://www.kaggle.com/datasets/uciml/iris

```python
# Importando bibliotecas para manipulação de dados
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Mostra os gráficos suprimindo mensagens
%matplotlib inline
```

```python
# Importando arquivo
df_iris = pd.read_csv('Iris.csv', index_col=0)
```

# Gráficos
Aqui iremos criar alguns gráficos para estudo de Análise de Dados.

```python
# GRÁFICO DA COLUNA - 'SepalLengthCm'
# Determinando o tamanho da figura
plt.figure(figsize=(15, 8))
df_iris['SepalLengthCm'].plot(lw=3)   # Escolhendo a coluna para mostrar no gráfico
plt.title('Gráfico 01', size=15)
```

```python
# GRÁFICO DA COLUNA - 'SepalWidthCm'
# Gráfico de histograma
plt.figure(figsize=(15, 8))
df_iris['SepalWidthCm'].plot(kind='hist', color='green')
plt.title('Gráfico 02', size=15)
```

```python
# GRÁFICO DA COLUNA - 'PetalLengthCm'
# Gráfico de box
plt.figure(figsize=(15, 8))
df_iris['PetalLengthCm'].plot(kind='box')
plt.title('Gráfico 03', size=15)
```

```python
# GRÁFICO DA COLUNA - 'PetalWidthCm'
# Histograma com densidade
plt.figure(figsize=(10, 8))
df_iris['PetalWidthCm'].plot(kind='density', color='red', lw=3)   # lw --> Espessura da linha
df_iris['PetalWidthCm'].plot(kind='hist', color='blue', density=True)
plt.title('Gráfico 04', size=15)
```

# Gráfico de dispersão

```python
# Selecionando duas colunas para apresentação
# SepalLengthCm --> Coluna colocada no eixo X
# PetalLengthCm --> Coluna colocada no eixo Y
# scatter --> Gráfico de dispersão
df_iris.plot(x='SepalLengthCm', y='PetalLengthCm', kind='scatter', color='purple')
plt.title('Dispersão das flores', size=15)
plt.show()
```

```python
# Gráfico de compartimento hexágonal
df_iris = pd.DataFrame(np.random.randn(1000, 2), columns=["PetalWidthCm", "PetalLengthCm"])
df_iris["PetalLengthCm"] = df_iris["PetalLengthCm"] + np.arange(1000)
df_iris.plot.hexbin(x="PetalWidthCm", y="PetalLengthCm", gridsize=25);
```

# Gráficos Lineares, Quadráticos e Cúbicos

```python
# Gráfico Linear
# Cria 100 números que vão de 0 até 2
x = np.linspace(0, 2, 100)
plt.plot(x, x)
```

```python
# Gráfico Quadrático
x = np.linspace(-2, 0, 100)
plt.plot(x, x**2)   # Elevado ao quadrado para criar um gráfico quadrático
```

```python
# Gráfico Cúbico
x = np.linspace(-2, 2, 100)
plt.plot(x, x**3)   # Elevado ao cubo para criar gráfico cúbico
```

```python
# Gráfico á Quarta Potência
x = np.linspace(-2, 2, 100)
plt.plot(x, x**4)   # Elevado a quarta potência
```

# Juntando os gráficos

```python
# Irei criar um gráfico com 3 gráficos
# 1° - Linear
# 2° - Quadrático
# 3° - Cúbico
x = np.linspace(-2, 1, 100)   # Especificação que será percorrida
plt.plot(x, x)   # Linear
plt.plot(x, x**2)   # Quadrático
plt.plot(x, x**3)   # Cúbico

plt.show()
```

# Adicionando algumas perfumarias
As perfumarias serão adicionadas por partes.

```python
# Adicionando Legenda e Título
x = np.linspace(-2, 1, 100)
plt.plot(x, x, label='Linear')   # label --> Colocando o nome da legenda
plt.plot(x, x**2, label='Quadrático')
plt.plot(x, x**3, label='Cúbico')

plt.title('Apresentação', size=15)   # Título do gráfico
plt.legend(fontsize='x-large')
plt.show()
```

```python
# Adicionando um 'GRID'
x = np.linspace(-2, 1, 100)
plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Quadrático')
plt.plot(x, x**3, label='Cúbico')

plt.title('Apresentação', size=15)
plt.legend(fontsize='x-large')
plt.grid(True)   # Adiciona uma grade no gráfico
plt.show()
```

```python
# Adicionando um título no eixo X e Y
x = np.linspace(-2, 1, 100)
plt.plot(x, x, label='Linear')
plt.plot(x, x**2, label='Quadrático')
plt.plot(x, x**3, label='Cúbico')

plt.title('Apresentação', size=15)
plt.legend(fontsize='x-large')
plt.xlabel('Eixo X', size=12)   # Adicionando eixo X
plt.ylabel('Eixo Y', size=12)   # Adicionando eixo Y
plt.grid(True)
plt.show()
```

# Personalizando os Labels

```python
x = np.linspace(-2, 1, 100)
plt.plot(x, x, label='Linear', color='purple')   # Adicionando cor com 'color'
plt.plot(x, x**2, label='Quadrático', color='blue')   # Adicionando cor com 'color'
plt.plot(x, x**3, label='Cúbico', color='green')   # Adicionando cor com 'color'

plt.title('Apresentação Personalizada', style='italic', size=15)   # Adicionando estilo 'style'
plt.legend(fontsize='x-large')
plt.xlabel('Eixo X', style='italic', size=12)   # Adicionando estilo 'style'
plt.ylabel('Eixo Y', style='oblique', size=12)   # Adicionando estilo 'style'
plt.grid(True)
plt.show()
```

# Mútiplos plots

[Documentação](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html#matplotlib.pyplot.subplot)

`fig, axes = plt.subplot(M, N)` 

* M são linhas
* N são colunas

```python
fig, axes = plt.subplots(1, 3, sharey='row')
x = np.linspace(0, 2, 100)

axes[0].plot(x, x, color='red')
axes[0].set_title('Linear', size=12)
axes[0].set_xlabel('Eixo X')
axes[0].set_ylabel('Eixo Y')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, sharey='row')
x = np.linspace(0, 2, 100)

# Gráfico Linear
axes[0].plot(x, x, color='red')
axes[0].set_title('Linear')
axes[0].set_xlabel('Eixo X')
axes[0].set_ylabel('Eixo Y')

# Gráfico Quadrático
axes[1].plot(x, x**2, color='blue')
axes[1].set_title('Quadrática')
axes[1].set_xlabel('Eixo X')

# Gráfico Cúbico
axes[2].plot(x, x**3, color='green')
axes[2].set_title('Cúbica')
axes[2].set_xlabel('Eixo X')
plt.show()
```
